```python
import os
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
from PIL import Image
from keras.models import load_model
```

```python
p1 = 'D:/plantvillage/raw/color'
classes = os.listdir(p1)
d = {}
l = []
cl = []
for c in classes:
    d[c] = len(os.listdir(os.path.join(p1,c)))
    l.append(d[c])
    cl.append(c)
    
print(d)
index = np.arange(len(l))
    
plt.bar(index,l)
plt.title('count of images in all categories')
plt.ylabel('count')
plt.xlabel('species index')
plt.xticks(index,cl,fontsize=10,rotation=90)
```
